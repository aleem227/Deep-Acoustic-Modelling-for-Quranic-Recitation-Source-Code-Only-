```python
import numpy 


def shuffle_in_unison(a, b):
    assert len(a) == len(b)
    shuffled_a = numpy.empty(a.shape, dtype=a.dtype)
    shuffled_b = numpy.empty(b.shape, dtype=b.dtype)
    permutation = numpy.random.permutation(len(a))
    for old_index, new_index in enumerate(permutation):
        shuffled_a[new_index] = a[old_index]
        shuffled_b[new_index] = b[old_index]
    return shuffled_a, shuffled_b
```

```python
# CNN2D_Model = Sequential()
    
# CNN2D_Model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(630, 40, 1)))
# CNN2D_Model.add(MaxPooling2D(pool_size=(2, 2)))
# CNN2D_Model.add(LSTM(none , return_sequences=True))
# CNN2D_Model.add(Flatten())
# CNN2D_Model.add(Dense(128, activation='relu'))
# CNN2D_Model.add(Dense(10, activation='softmax'))

# CNN2D_Model.summary()


CNN2D_Model = Sequential()
CNN2D_Model.add(
    Conv2D(64, (3, 3), padding="same", activation="tanh", input_shape=(704, 40, 1), kernel_regularizer=L2(0.01))
)
# CNN2D_Model.add(Flatten())
CNN2D_Model.add(MaxPool2D(pool_size=(2, 2)))
CNN2D_Model.add(Conv2D(64, (3, 3), padding="same", activation="tanh"))
CNN2D_Model.add(MaxPool2D(pool_size=(2, 2)))
CNN2D_Model.add(Conv2D(64, (3, 3), padding="same", activation="tanh"))

# # CNN2D_Model.add(MaxPool2D(pool_size=(2, 2)))
# CNN2D_Model.add(Conv2D(64, (3, 3), padding="same", activation="relu"))
# # CNN2D_Model.add(MaxPool2D(pool_size=(2, 2)))
# CNN2D_Model.add(Conv2D(64, (3, 3), padding="same", activation="relu"))

CNN2D_Model.add(MaxPool2D(pool_size=(2, 2)))
CNN2D_Model.add(Dropout(0.1))
CNN2D_Model.add(Flatten())
CNN2D_Model.add(Dense(1024, activation="tanh"))
CNN2D_Model.add(Dense(10, activation="softmax"))
CNN2D_Model.summary()
```

```python
CNN2D_Model.compile(
    optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"]
)
```

```python
t0 = time.time()

CNN2D_Results = CNN2D_Model.fit(
    X_train, y_train, epochs=20, batch_size=64, validation_data=(X_test, y_test)
)

```
