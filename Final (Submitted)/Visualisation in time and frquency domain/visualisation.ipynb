```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Load the WAV file using librosa
file_path = '105-al-fil.wav'
audio, sr = librosa.load(file_path, sr=16000)

# Visualize in the time domain
plt.figure(figsize=(8, 4))
plt.plot(np.arange(len(audio)) / sr, audio)
plt.title('Surah Al-Maun - Sheikh Saad Al Ghamdi')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Perform the Fourier Transform to obtain the frequency domain representation
audio_fft = np.fft.fft(audio)
magnitude = np.abs(audio_fft)
frequency = np.linspace(0, sr, len(magnitude))

# Visualize in the frequency domain
plt.figure(figsize=(8, 4))
plt.plot(frequency[:len(magnitude)//2], magnitude[:len(magnitude)//2])
plt.title('Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
mfcc = librosa.feature.mfcc(audio, sr=sr, n_mfcc=40)

# Plot the MFCC graph
plt.figure(figsize=(8, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC of Surah Al-Fil', fontweight='bold')  # Make the title bold
plt.xlabel('Time (s)', fontweight='bold')  # Make the x-label bold
plt.ylabel('MFCC Coefficients', fontweight='bold')  # Make the y-label bold
plt.show()
```
